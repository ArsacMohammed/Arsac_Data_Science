```python
plt.figure(figsize=(5,3),dpi=100) # for graph picture quality
x=[0,1,2,3,4,5]
y=[0,2,4,6,8,10]
plt.xticks([0,1,2,3])
#plt.yticks([0,2,4,6])
plt.plot(x,y,'r.--',label="2x")
x2=np.arange(0,4.5,0.5)  # the first two numbers are range and the last number is fixed size 
plt.plot(x2,x2**2,'g.--',markersize=10,label="x^2")# i have used the short notation
plt.title("Our First Graph",{'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(5,3),dpi=100) 
x=[0,1,2,3,4,5]
y=[0,2,4,6,8,10]
plt.xticks([0,1,2,3])
#plt.yticks([0,2,4,6])
plt.plot(x,y,'r.--',label="2x")
x2=np.arange(0,4.5,0.5)  
plt.plot(x2[0:6],x2[0:6]**2,'g.--',markersize=10,label="x^2")
plt.plot(x2[5:],x2[5:]**2,'g',markersize=10,label="x^2")
plt.title("Our First Graph",{'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.savefig('mygraph.png',dpi=200)
plt.show()
```

```python
labels=['A','B','C']
values=[1,2,5]
bars=plt.bar(labels,values)
patterns=['/','o','*']
for i in range(0,3,1):
    bars[i].set_hatch(patterns[i])
plt.figure(figsize=(5,3),dpi=100)
plt.show()
```
